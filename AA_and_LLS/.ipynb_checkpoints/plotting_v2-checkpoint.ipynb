```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import os
import glob
import matplotlib.pyplot as plt
#import cartopy.crs as ccrs
#import cartopy.feature as cfeature
import pandas as pd
import json
import cftime
#from itertools import product
from cftime import DatetimeNoLeap
#import Gridding
#from nc_processing import *
#from JASMIN_utils import *
#from analysis import * 
#from plotting import *
#import esmvalcore.preprocessor
import xesmf as xe
import warnings
%matplotlib inline
import seaborn as sns
#sns.set()
from tqdm import tqdm
from scipy.stats import linregress
from nc_processing import calc_spatial_mean
from xmip.preprocessing import rename_cmip6

def quick_scat(df, x_var, y_var, save=False, out_name='fig.png'):
    plt.subplots(figsize=(8,6))
    regression = linregress(df[x_var],df[y_var])
    sns.regplot(data= df, x=x_var, y=y_var, scatter=False)
    sns.scatterplot(data= df, x=x_var, y=y_var)

    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.title('slope = {r}, pvalue = {p}'.format(r=np.round(regression[0],2), p=np.round(regression[3],2)))
    if save:  
        plt.savefig(outname, dpi=300)
    
```

```python
ds_list = []

for x in os.listdir('int_ncs/'):
    ds = xr.open_dataset('int_ncs/'+x)
    ds_list.append(ds)

ds = xr.concat(ds_list, dim='Model')

broken_mods = ['E3SM-1-1', 'EC-Earth3-Veg-LR', 'ACCESS-ESM1-5', 'FGOALS-g3']

ds = ds.drop(labels=broken_mods, dim='Model')

models = ds.Model.values
```

```python
ds
```

```python

Arctic_warming_rates = []
Tropics_warming_rates = []
for mod in models:
    ds_m = ds.sel(Model=mod)
    ds_m = ds_m.sel(year=slice(2000, 2050))
    r_a = linregress(ds_m['year'], ds_m['Arctic_66_tas']).slope
    r_t = linregress(ds_m['year'], ds_m['Tropics_-30_30_tas']).slope
    Arctic_warming_rates.append(r_a*10)
    Tropics_warming_rates.append(r_t*10)
    
Baseline_lls = []
for mod in models:
    ds_m = ds.sel(Model=mod)
    ds_m = ds_m.sel(year=slice(1850, 1900)).mean(dim='year')
    Baseline_lls.append(ds_m.LLS_75.item())

Late_cent_A_warmings = []
for mod in models:
    ds_m = ds.sel(Model=mod)
    ds_m_l = ds_m.sel(year=slice(2050, 2100)).mean(dim='year')
    ds_m_b = ds_m.sel(year=slice(1850, 1900)).mean(dim='year')
    Late_cent_A_warmings.append(ds_m_l.Arctic_66_tas.item()-ds_m_b.Arctic_66_tas.item())

df1 = pd.DataFrame({'Model':models,
                    'Arctic_warming_rate_C_per_decade_66N':Arctic_warming_rates,
                    'Tropics_warming_rate_C_per_decade':Tropics_warming_rates,
                    'Baseline_LLS_75N':Baseline_lls,
                    'Late_cent_Arctic_warming_C':Late_cent_A_warmings})

df1['AA'] = df1['Arctic_warming_rate_C_per_decade_66N']/df1['Tropics_warming_rate_C_per_decade']
#df1
```

```python
### better plot
import matplotlib
matplotlib.rcParams.update({'font.size': 16})

fig, ax = plt.subplots(figsize=(16, 8))


x, y = df1['Baseline_LLS_75N'], df1['AA']
ax.scatter(x, y, color='black')
#ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), c='gray')
print(linregress(x, y))

sns.regplot(ax=ax, data= df1,
            x='Baseline_LLS_75N', y='AA',
            scatter=False, color='gray', ci=90)

ax.set_xlabel('Pre-ind. low-level stability (°C)')
ax.set_ylabel('Arctic amplification')
    
plt.savefig('Figures/LLS_vs_AA_scatter.png', dpi=350)
plt.show()
```

```python
DF = pd.DataFrame()
window = 10

for mod in models:
    data = ds.sel(Model=mod)
    plt.plot(data.year.values, 
             data.Arctic_66_tas.values-273.15, 
             label=mod)
    df = pd.DataFrame({'Model':np.repeat(mod, len(data.year.values)),
                       'Year':data.year.values,
                       'Arctic_66_tas':data.Arctic_66_tas.values-273.15,
                       'LLS_75':data.LLS_75.values})
   
    df['Arctic_66_tas_rolling'] = df['Arctic_66_tas'].rolling(window=window, center=True).mean()
    df['LLS_75_rolling'] = df['LLS_75'].rolling(window=window, center=True).mean()
    
    DF = pd.concat([DF, df])
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(14, 6), sharey=False)

ax1 = axs[0]
var1 = 'Arctic_66_tas_rolling'

means = DF.groupby('Year')[var1].mean().dropna()
ax1.plot(means.index.values,means.values,lw=1.5,color='black')

quantiles=[0.05, 0.95]
bottoms = DF.groupby('Year')[var1].quantile(q=quantiles[0]).dropna()
tops = DF.groupby('Year')[var1].quantile(q=quantiles[1]).dropna()

ax1.fill_between(means.index.values, bottoms.values, tops.values, color='gray', alpha=0.3)
ax1.set_title('(a)', loc='left')
ax1.grid(which='major', axis='both')
ax1.set_ylabel('Temperature (°C)')

#ax1.legend(fontsize='large', loc='upper left')

ax2 = axs[1]
var2 = 'LLS_75_rolling'
means = DF.groupby('Year')[var2].mean().dropna()
ax2.plot(means.index.values,means.values,lw=1.5,color='black')

quantiles=[0.10, 0.90]
bottoms = DF.groupby('Year')[var2].quantile(q=quantiles[0]).dropna()
tops = DF.groupby('Year')[var2].quantile(q=quantiles[1]).dropna()

ax2.fill_between(means.index.values, bottoms.values, tops.values, color='gray', alpha=0.3)
#ax2.axhline(0)
ax2.set_title('(b)', loc='left')
ax2.grid(which='major', axis='both')
ax2.set_ylabel('LLS (°C)')

#ax1.legend(fontsize='large', loc='upper left')
fig.supxlabel('Year')
plt.tight_layout()
plt.savefig('Figures/timeseries_warming_LLS.png', dpi=350)
```

```python
fig, axs = plt.subplots(nrows=5, ncols=8, figsize=(22,16), sharex=True, sharey=True)

i=0
rs = []
ps = []
for mod in models:
    ds_m = ds.sel(Model=mod)
    ax = axs.flatten()[i]
    ax.scatter(ds_m.Arctic_66_tas.values-273.15, ds_m.LLS_75.values, s=1)
    r = linregress(ds_m.Arctic_66_tas.values-273.15, ds_m.LLS_75.values).slope
    p = linregress(ds_m.Arctic_66_tas.values-273.15, ds_m.LLS_75.values).pvalue
    rs.append(r)
    ps.append(p)
    i=i+1
    ax.set_title(mod)
R = np.mean(rs)
for p in ps:
    if p >0.01:
        print(p)
R_std = np.std(rs)
print(np.round(R, 2), '+-', np.round(R_std, 2))
fig.supxlabel('DJF Arctic 2m air temperature')
fig.supylabel('DJF High Arctic LLS')
plt.tight_layout()
plt.savefig('Figures/LLS_AWarming_ind_mods.png', dpi=350)
plt.show()
```
